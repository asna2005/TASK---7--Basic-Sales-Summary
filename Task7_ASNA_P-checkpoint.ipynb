```python
import sqlite3

# reconnect to the database
conn = sqlite3.connect("sales_data.db")
cursor = conn.cursor()

# run queries again
cursor.execute("SELECT * FROM sales")
rows = cursor.fetchall()
print(rows)
```

#### **Create the sales table**

```python
#
cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    product TEXT,
    quantity INTEGER,
    price REAL
)
""")
```

#### **Insert Data**

```python

data = [
    ('Laptop', 5, 50000),
    ('Phone', 8, 20000),
    ('Headphones', 10, 1500),
    ('Tablet', 3, 30000),
    ('Charger', 15, 800)
]
cursor.executemany("INSERT INTO sales VALUES (?, ?, ?)", data)
conn.commit()
print("âœ… Sample data inserted successfully!")

```

```python

query = """
SELECT 
    product,
    SUM(quantity) AS total_quantity,
    SUM(quantity * price) AS total_revenue
FROM sales
GROUP BY product
"""
df = pd.read_sql_query(query, conn)
print("\nðŸ“Š Basic Sales Summary:\n")
print(df)
```

#### **Total Quantity & Total Revenue**

```python

total_qty = df['total_quantity'].sum()
total_revenue = df['total_revenue'].sum()
print("\nâœ… Overall Summary:")
print(f"Total Quantity Sold: {total_qty}")
print(f"Total Revenue: â‚¹{total_revenue}")
```

#### **Revenue by Product**

```python

plt.figure(figsize=(8,5))
plt.bar(df['product'], df['total_revenue'], color='skyblue', edgecolor='black')
plt.title("Revenue by Product")
plt.xlabel("Product")
plt.ylabel("Revenue (â‚¹)")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

#### **Quantity Sold by Product**

```python
plt.figure(figsize = (8,5))
plt.bar(df['product'], df['total_quantity'], color = 'lightgreen' , edgecolor = 'black')
plt.title("Quantity Sold by Product")
plt.xlabel("product")
plt.ylabel("Quantity Sold")
plt.grid(axis ='y', linestyle ='--', alpha = 0.6)
plt.show()
```

#### **Quantity Vs Revenue**

```python
plt.figure(figsize = (7,5))

plt.scatter(df['total_quantity'], df['total_revenue'] , color = 'orange' ,  edgecolor = 'black' , s=100)

plt.title("Quantity Vs Revenue")

plt.xlabel("Total Quantity")

plt.ylabel("Total Revenue")

plt.grid(True , linestyle ='--' , alpha = 0.5)

plt.show()
```
